```python
sns.countplot(x='Product line',data=df,hue='Gender',order=df['Product line'].value_counts().index)
plt.xticks(rotation=85)
```

```python
city_count=df.groupby(['City','Product line'])['Total'].sum().reset_index()
city_count
```

```python
sns.barplot(x='Product line',y='Total',hue='City',data=city_count)
plt.xticks(rotation=85)
```

```python
payment_cnt=pd.DataFrame(df['Payment'].value_counts().reset_index())
payment_cnt
```

```python
sns.barplot(x='index',y='Payment',data=payment_cnt)
plt.xlabel('Payment Mode')
plt.ylabel('Count')
```
